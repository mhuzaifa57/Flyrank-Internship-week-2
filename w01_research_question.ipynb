## 2. The question: decision, action, cost of a wrong call


**Research question:** Given a page's current signals (age, staleness, visibility,
position, trend), can we rank pages by how urgently they need a content refresh —
better than a simple hand-written rule?

**Unit of analysis:** a single page (one row = one URL at one point in time).

**Output:** a ranked list (or score) of pages, from "refresh first" to "leave alone."

**The action:** an SEO/content team pulls the top N pages off this list into their
next sprint instead of guessing or working alphabetically / by traffic alone.

**Cost of a wrong call:**
- False positive (flagged "urgent" but fine): wastes a writer/editor's time on a
  page that didn't need it — a few hours of paid work, low but real cost.
- False negative (missed a page that's actually declining): a page keeps losing
  impressions/rank silently until the next audit cycle — lost traffic compounds
  for weeks before anyone notices.
Because false negatives compound silently, I'd rather the model be biased toward
over-flagging than under-flagging.

**Why data/ML helps at all:** a hand-written rule (e.g. "stale AND visible") can
only combine a couple of signals with fixed thresholds. Whether staleness matters
more at position 3 than position 30, or whether word count matters only for
certain content types, is an interaction a simple rule can't express but a model
can learn directly from data.

## 3. Quick look at the data (2-3 real numbers)

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import os, sys

# Walk up from wherever this notebook's kernel started until we find the repo root
# (identified by the presence of data/raw/) — needed because this notebook lives
# in work/notebooks/, not the repo root.
while not os.path.isdir("data/raw") and os.getcwd() != "/":
    os.chdir("..")

print("Working dir:", os.getcwd())
assert os.path.exists("data/raw/content_refresh_anonymized.csv"), (
    "Could not find the starter CSV — check you're running from inside the cloned repo."
)
import pandas as pd
import numpy as np

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

trend = df["trend_direction"].astype(str).str.strip().str.lower()
is_declining = trend.eq("down")

# Number 1
declining_rate = is_declining.mean()
print(f"Share of pages currently declining: {declining_rate:.1%}")

# Number 2
stale = df["days_since_last_update"].fillna(0) >= 180
decline_rate_stale     = is_declining[stale].mean()   if stale.any()    else float("nan")
decline_rate_not_stale = is_declining[~stale].mean()  if (~stale).any() else float("nan")
print(f"Decline rate, stale pages:     {decline_rate_stale:.1%}")
print(f"Decline rate, non-stale pages: {decline_rate_not_stale:.1%}")

# Number 3
declining_impressions = df.loc[is_declining, "impressions_90d"].sum()
total_impressions = df["impressions_90d"].sum()
if total_impressions > 0:
    print(f"Share of all impressions sitting on declining pages: {declining_impressions/total_impressions:.1%}")
else:
    print("No impressions data available to compute this share.")
```
